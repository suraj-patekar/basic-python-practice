```python
def factorial_iterative(n):
    """ parameter: n
    returns: n*(n-1)*(n-2)*(n-3)*......*1"""
    fac = 1
    for i in range(n):
        fac *= (i+1)
    return fac
```

```python
factorial_iterative(5)
```

```python
def factorial_recursive(n):
    """ parameter: n
    returns: n*(n-1)*(n-2)*(n-3)*......*1"""
    if n==0 or n==1:
        return 1
    elif n < 0:
        return "Factorial of negative number does not exist"
    else:
        return n * factorial_recursive(n-1)
```

```python
factorial_recursive(5)
```

```python
factorial_recursive(10)/factorial_recursive(5)
```

```python
n = int(input("number to find factorial: "))
```

```python
factorial_recursive(n)
```

```python
def fibbo(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return (fibbo(n-1) + fibbo(n-2))
```

```python
fibbo(10)
```

```python
def recur_fibo(n):
    if n <= 1:
        return n
    else:
        return(recur_fibo(n-1) + recur_fibo(n-2))
```

```python
recur_fibo(10)
```

```python
def dec(func):
    def exec_now():
        print("Executing now")
        func()
        print("Executed successfully")
    return exec_now()
        
```

```python
def who_is_suraj():
    print("Suraj is an Associate Data Engineer Trainee")
    
```

```python
who_is_suraj()
```

```python
who_is_suraj = dec(who_is_suraj)
```

```python
@dec
def who_is_suraj():
    print("Suraj is an Associate Data Engineer Trainee")
```

```python
class Employee:
    no_of_leaves = 10
    def __init__(self,name,salary,role):  #constructor to provide arguments in the class
        self.name = name
        self.salary = salary
        self.role = role
        
    
    def details(self):
        return f"Name is {self.name}, salary is {self.salary} and job title is {self.role}"
    
    @classmethod   #changes attribute of class not of instance
    def change_leaves(cls,new_leaves):
        cls.no_of_leaves = new_leaves
        
    @classmethod
    def from_dash(cls,string):
        return cls(*string.split("-"))
    
    @staticmethod
    def printgood(string):
        print("this is good", string)
```

```python
Employee.printgood("Suraj")
```

```python
harry = Employee("Harry",5000,"Instructor")
suraj = Employee("Suraj",2000,"Trainee")   #These arguements go into __init__ in Class
karan = Employee.from_dash("Karan-3000-Associate")
```
