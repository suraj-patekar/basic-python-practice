#### Write a Pandas program to split a given dataset, group by two columns and convert other columns of the dataframe into a dictionary with column header as key

```python
df = pd.DataFrame({'school_code': ['s001','s002','s003','s001','s002','s004'],
                  'class':['V','V','VI','VI','V','VI'],
                  'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes','Eesha Hinton','Gino Mcneill','David Parkes'],
                  'date_of_birth':['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
                  'age':[12,12,13,13,12,14],
                   'height':[173,192,186,167,151,159],
                   'weight':[35,32,33,30,31,32],
                   'address':['street1','street2','street3','street1','street2','street4']
                  })
```

```python
 grouped_object = df.groupby(['school_code','class'])
```

```python
dict_data = []

for g, d in grouped_object:
    group = dict(zip(['school_code','class'], g))
    other_columns = []
    for _, data in d.iterrows():
        data = data.drop(labels=['school_code','class'])
        other_columns.append(data.to_dict())
    group['other_columns'] = other_columns
    dict_data.append(group)
```

```python
dict_data
```
